```python
import pandas as pd
lgb =  pd.read_csv("submit.csv")
xgb =  pd.read_csv("submit_XGBoost.csv")
cat = pd.read_csv("cat_submit.csv")
```

```python

df_all = lgb[['kind', 'date']].copy()
df_all['lgb'] = lgb['mode_price']
df_all['xgb'] = xgb['mode_price']
df_all['cat'] = cat['mode_price']

df_all['mean_pred'] = df_all[['lgb', 'xgb', 'cat']].mean(axis=1)

df_all[["kind", "date", "mean_pred"]].rename(
    columns={"mean_pred": "mode_price"}
).to_csv("submission.csv", index=False)
```
